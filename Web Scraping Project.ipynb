```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
## store the website in a variable 
website = 'https://mayxachtay.vn/Laptop-xach-tay-00112.html'
```

```python
## get request
response = requests.get(website)
```

```python
## status code
response.status_code
```

Soup object

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

Result

```python
result = soup.find_all('div', {'class':'bginfodiv'})
```

```python
## number of products in the page 
len(result)
```

```python
result[0]
```

Get necessary information of products

```python
## Info (Name, Graphic, ... )
result[0].find('a',{'class':'bginfo'}).get_text()
```

```python
## Price
result[0].find('span',{'class':'priceinfo'}).get_text()
```

Extract product link

```python
result[0].find('a',{'class':'bginfo'}).get('href')
```

```python
## Get product details 
result[0].find('a',{'class':'bginfo'}).get_text().strip().replace('\n',',').replace('\r',',')
```

Put everything together inside a For-loop

```python
product_name = []
product_price = []
product_link = []
for a in result: 
    #Info
    try:
        product_name.append(a.find('a',{'class':'bginfo'}).get_text().strip().replace('\n',',').replace('\r',','))
    except:
        product_name.append('n/a')
    #Price
    try:
        product_price.append(a.find('span',{'class':'priceinfo'}).get_text())
    except:
        product_price.append('n/a')
    #Link
    try:
        product_link.append(a.find('a',{'class':'bginfo'}).get('href'))
    except:
        product_link.append('n/a')
```

```python
## Check how many products in the page 
len(product_link)
```

Create a dataframe

```python
product = pd.DataFrame({'Name':product_name,'Price':product_price,'Link':product_link})
product
```

```python
## Output to excel
product.to_excel('abcd.xlsx',index=False)
```
